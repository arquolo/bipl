```python
from itertools import product
from dataclasses import dataclass


@dataclass
class A:
    shape: tuple[int, ...]
    tile_shape: tuple[int, ...]

    def _make_index(self, box: np.ndarray):
        # box: (yx lo/hi)
        # (yx)
        min_ = box[:, 0]
        box_size = box @ [-1, 1]
        # (yx lo/hi)
        vbox = box.T.clip(0, self.shape).T  # box within slide
        o_loc = vbox - min_[:, None]  # region of `out` to fill

        # (yx)
        vbox_size = vbox @ [-1, 1]
        # (yx lo/hi)
        o_loc = o_loc.clip(0, box_size[:, None])

        if not vbox_size.all():
            return [], o_loc

        # (yx)
        tile_size = self.tile_shape
        i_lo = vbox[:, 0] // tile_size  # i_lo * tile <= vmin
        i_hi = -(-vbox[:, 1] // tile_size)  # i_hi * tile >= vmax

        # (ny) (nx)
        (*ids,) = map(range, i_lo, i_hi)
        oxs: list[list[slice]] = []
        txs: list[list[slice]] = []
        for i_lo_, i_hi_, tsize, vspan, min__ in zip(
            i_lo, i_hi, tile_size, vbox, min_
        ):
            # (n + 1)
            ts = np.arange(i_lo_, i_hi_ + 1) * tsize
            vts = ts.clip(*vspan)
            vts_src = vts - min__

            # (n), result slices
            (*o_crops,) = map(slice, vts_src[:-1], vts_src[1:])
            oxs.append(o_crops)

            # (n), tile slices
            (*t_crops,) = map(slice, vts[:-1] - ts[:-1], vts[1:] - ts[:-1])
            txs.append(t_crops)

        return (
            [*zip(product(*ids), product(*oxs), product(*txs), strict=True)],
            o_loc,
        )


a = A(shape=(400,), tile_shape=(256,))
for p0, p1 in [
    [-100, -50],
    [-100, -25],
    [-100, 0],
    [-100, +25],
    [-100, +500],
    [+375, +500],
    [+400, +500],
    [+425, +500],
    [+450, +500],
]:
    coords, [[e0, e1]] = a._make_index(np.array([[p0, p1]]))
    print(f'\nbox = [{p0}:{p1}], roi = [{e0}:{e1}]')
    if coords:
        print(*coords, sep='\n')
```

```python
def hist(svhwc: np.ndarray) -> np.ndarray:
    s, v, _, _, _ = svhwc.shape
    im = rearrange(svhwc, 's v h w c -> (s v) (h w) c')

    k = 4
    im = cv2.cvtColor(im, cv2.COLOR_RGB2LAB)
    im //= 256 // k  # (s v) n 3

    q = im.astype('u4') @ [k * k, k, 1]  # (s v) n
    q += (np.arange(s * v) * k * k * k)[:, None]

    h = np.bincount(q.ravel(), minlength=s * v * k * k * k).astype('f4')
    sv444 = h.reshape(s, v, k, k, k)
    sv444 /= sv444.sum((2, 3, 4), keepdims=True)  # L1 norm
    return sv444


def hsa_op(sv444: np.ndarray) -> np.ndarray:
    # Mean variance
    return sv444.var(0).sum((1, 2, 3))  # (1+variants)
```

```python
import numpy as np
from glow import pascal
import pandas as pd

p_fail = 0.05

rows = []
for parity, stripes, blocks, level in [
    (7, 8, 1, '1'),
    (2, 4, 2, '60'),
    (2, 8, 1, '6'),
    (1, 2, 4, '10'),
    (1, 4, 2, '50'),
    (1, 8, 1, '5'),
    (0, 8, 1, '0'),
    (0, 1, 1, 'single drive'),
]:
    data_pct = (stripes - parity) * blocks

    p_loss = (
        (p_fail / (1 - p_fail)) ** np.arange(parity + 1, stripes + 1)
        @ pascal(stripes + 1)[parity + 1 :]
    ) * (1 - p_fail) ** stripes

    p_loss = 1 - (1 - p_loss) ** blocks
    max_fail = parity * blocks

    read_perf = stripes * blocks
    write_perf = (stripes - parity) * blocks

    rows.append(
        {
            'level': level,
            'TB': data_pct * 2,
            'read perf': read_perf,
            'write perf': write_perf,
            'loss %': round(p_loss, 4) * 100,
            'max fail': max_fail,
        }
    )
pd.DataFrame.from_records(rows).set_index('level')
```
